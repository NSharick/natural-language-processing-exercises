```python
def parse_people(person):
    output = {}
    output['name'] = person.find('h2').text
    output['statement'], output['email'], output['phone'], output['address'] = [p.text.strip() for p in person.find_all('p')]
    return output
    
    
```

```python
pd.DataFrame([parse_people(person) for person in people])
```

1. Codeup Blog Articles

- Visit Codeup's Blog and record the urls for at least 5 distinct blog posts. For each post, you should scrape at least the post's title and content.

- Encapsulate your work in a function named get_blog_articles that will return a list of dictionaries, with each dictionary representing one article. The shape of each dictionary should look like this:

    - {
    'title': 'the title of the article',
    'content': 'the full text content of the article'
}

```python
url = 'https://codeup.com/blog/'
headers = {'user-agent': 'Innis Data Science Cohort'}
response = get(url, headers=headers)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
articles = soup.select('article.et_pb_post')
```

```python
article = articles[0]
article
```

```python
len(articles)
```

```python
article.find('h2').text
```

```python
[p.text for p in article.find_all('p')]
```

```python
def parse_blog(article):
    output = {}
    output['title'] = article.find('h2').text
    output['date_published'] = article.select_one('span.published').text
    output['content'] = article.select_one('div.post-content-inner').text
    return output
```

```python
parse_blog(articles[6])
```

```python
pd.DataFrame([parse_blog(article) for article in articles])
```

2. News Articles

- We will now be scraping text data from inshorts, a website that provides a brief overview of many different topics.

- Write a function that scrapes the news articles for the following topics:

    - Business
    - Sports
    - Technology
    - Entertainment

- The end product of this should be a function named get_news_articles that returns a list of dictionaries, where each dictionary has this shape:


  - {
    'title': 'The article title',
    'content': 'The article content',
    'category': 'business' # for example
}

**InShorts Blog**

```python
url = 'https://blog.inshorts.com/'
response = get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
posts = soup.select('article.post-excerpt')
post = posts[0]
post
```

```python
post.find('h2').text
```

```python
post.find('p').text
```

```python
def parse_posts(post):
    output = {}
    output['title'] = post.find('h2').text
    output['content'] = post.find('p').text
    return output
```

```python
parse_posts(posts[0])
```

```python
len(posts)
```

```python
pd.DataFrame(parse_posts(post) for post in posts)
```

**InShorts Business**

```python
url = 'https://inshorts.com/en/read/business'
response = get(url)
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
soup
```

```python
articles = soup.select('div.news-card')
article = articles[0]
article
```

```python
len(articles)
```

```python
article.select_one('div.news-card-title').text
```

```python
article.select_one('div.news-card-content').text
```

```python
def parse_business_news(article):
    output = {}
    output['title'] = article.select_one('div.news-card-title').text.strip()
    output['content'] = article.select_one('div.news-card-content').text.strip()
    output['category'] = 'business'
    return output
```

```python
parse_business_news(article)
```

```python
business_articles = pd.DataFrame([parse_business_news(article) for article in articles])
```

```python
business_articles.head()
```
